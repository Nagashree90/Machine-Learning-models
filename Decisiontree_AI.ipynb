### What is Encoding?

In machine learning, **encoding** refers to the process of converting data from one format to another, typically to make it suitable for a machine learning algorithm. Many algorithms, especially those based on mathematical equations (like regression or support vector machines), cannot directly work with non-numeric data (e.g., text, categories). Therefore, we need to convert these non-numeric inputs into a numerical representation.

For example, our `y` DataFrame currently contains target labels as integers (0, 1, 2) representing different Iris species. If these were strings like 'setosa', 'versicolor', 'virginica', we would need to encode them. Even when they are integers, depending on the model, we might choose a different encoding method.

There are several types of encoding, but for **categorical data** (data that can be divided into groups or categories), two very common methods are Label Encoding and One-Hot Encoding.

#### 1. Label Encoding

**Label Encoding** assigns a unique integer to each category based on its alphabetical order or the order in which it appears. For example, if you have categories 'Red', 'Green', 'Blue', they might be encoded as 0, 1, 2 respectively.

**Pros:** Simple to implement, reduces dimensionality.
**Cons:** Implies an ordinal relationship between categories (e.g., 2 > 1 > 0), which may not exist. This can mislead models into thinking that higher numbers are 'better' or more 'important', potentially hurting performance if the categories are nominal (no inherent order).

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

# Create a sample DataFrame with categorical data (similar to our 'y' if it had string labels)
# For demonstration, let's assume 'y' had string labels first
sample_species = pd.DataFrame({
    'species': ['setosa', 'versicolor', 'virginica', 'setosa', 'versicolor']
})

print("Original Sample Species:")
display(sample_species)

# Apply Label Encoding
label_encoder = LabelEncoder()
sample_species['species_encoded_label'] = label_encoder.fit_transform(sample_species['species'])

print("\nSample Species after Label Encoding:")
display(sample_species)

print("\nMapping of labels to integers:")
for i, name in enumerate(label_encoder.classes_):
    print(f"{name}: {i}")

# Our 'y' DataFrame already has label-encoded integers (0, 1, 2) directly from load_iris()
print("\nOur 'y' DataFrame already represents a form of Label Encoding:")
display(y.head())
```

```python
from sklearn.preprocessing import OrdinalEncoder
import pandas as pd

# We'll use the 'sample_species' DataFrame we created for Label Encoding demonstration
# Original Sample Species:
print("Original Sample Species:")
display(sample_species)

# Initialize OrdinalEncoder
# By default, OrdinalEncoder assigns integer categories based on the order of unique values it encounters
ordinal_encoder = OrdinalEncoder()

# Fit and transform the 'species' column. OrdinalEncoder expects a 2D array.
sample_species['species_encoded_ordinal'] = ordinal_encoder.fit_transform(sample_species[['species']])

print("\nSample Species after Ordinal Encoding:")
display(sample_species)

print("\nMapping of labels to integers by OrdinalEncoder:")
# The categories_ attribute stores the unique categories found for each feature
for i, category_list in enumerate(ordinal_encoder.categories_):
    print(f"Feature {i} mapping: {dict(zip(category_list, range(len(category_list))))}")

print("\nNote: For this dataset and without specifying order, LabelEncoder and OrdinalEncoder yield similar results.")
print("Our 'y' DataFrame already contains these integer labels from `load_iris()` and can be considered ordinally encoded.")
```

#### 2. One-Hot Encoding

**One-Hot Encoding** creates new binary (0 or 1) columns for each unique category present in the feature. If a data point belongs to a particular category, the corresponding column will have a '1' and all other new columns for that feature will have '0'.

**Pros:** Does not imply any ordinal relationship, making it suitable for nominal categorical data. Many models prefer this representation.
**Cons:** Increases the dimensionality of the dataset significantly, especially if there are many unique categories. This can lead to the 'curse of dimensionality' and increased computational cost.

```python
from sklearn.preprocessing import OneHotEncoder

# To apply One-Hot Encoding, let's first convert our 'y' (integer labels) back to string names
# for a clearer demonstration of the process, then encode them.
# In a real scenario, you'd apply this directly to string categorical data.

# Get original species names
species_names = iris.target_names

# Map integer labels in 'y' to original species names
y_nominal = y['species'].map(lambda x: species_names[x])
y_nominal_df = y_nominal.to_frame(name='species_name')

print("Original 'y' (nominal representation):")
display(y_nominal_df.head())

# Apply One-Hot Encoding
# OneHotEncoder expects a 2D array, so we reshape y_nominal_df[['species_name']]
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
encoded_data = encoder.fit_transform(y_nominal_df[['species_name']])

# Create a DataFrame from the one-hot encoded data
y_one_hot = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(['species_name']))

print("\n'y' after One-Hot Encoding:")
display(y_one_hot.head())

# You can see how each species now has its own binary column.
```

### When to use which encoding?

-   **Label Encoding** is generally suitable for **ordinal categorical data** (where categories have a meaningful order, e.g., 'Small', 'Medium', 'Large'). It can also be used for target variables (like our `y` for classification) where the model implicitly handles the integer labels without assuming an order.
-   **One-Hot Encoding** is preferred for **nominal categorical data** (where categories have no inherent order, e.g., 'Red', 'Blue', 'Green'). It prevents the model from misinterpreting arbitrary numerical relationships between categories.
